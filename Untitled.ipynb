```python
class MyTwythonStreamer(TwythonStreamer):
    def on_success(self, data):
        if data['lang'] == 'en':
            tweets.append(data)
            print('Recieved tweet #{}'.format(len(tweets)))
        if len(tweets)>= 50:
            self.disconnect()
    def on_error(self, status_code, data):
        print(status_code, data)
        self.disconnect()
        
```

```python
class MyTwythonStreamer(TwythonStreamer):
    def on_success(self, data):
        if data['lang'] == 'en' and data['retweeted'] == False:
            tweet_text = data['text']
            tweet_date = data['timestamp_ms']
            compound_sentiment = analyser.polarity_scores(tweet_text)['compound']
            name = data['user']['name']
            screen_name = data['user']['screen_name']
            twitter_bio = data['user']['description']
            follower_count = data['user']['followers_count']
            friends_count = data['user']['friends_count']
            statuses_count = data['user']['statuses_count']
            favourites_count = data['user']['favourites_count']
            profile_image_url = data['user']['profile_image_url_https']
            tweet_identity = data['id']
            q = 'INSERT INTO twitter.tweets_grabbed VALUES(?,?,?,?,?,?,?,?,?,?,?,?)'
            cur.execute(q, tweet_date, tweet_text, compound_sentiment, name, screen_name, twitter_bio, follower_count, friends_count, statuses_count, favourites_count,  profile_image_url, tweet_identity )
            cur.commit()
    def on_error(self, status_code, data):
        print(status_code, data)
        self.disconnect()
```
